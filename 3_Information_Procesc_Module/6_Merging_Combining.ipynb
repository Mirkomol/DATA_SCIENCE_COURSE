# More Example

```python
A = pd.DataFrame({'Transport':['poeyezd','avtomobil','kema'],
                 'Tezligi':['70','80','90']})
```

```python
A
```

```python
B = pd.DataFrame({'Transport':['poeyezd','avtomobil','tramvay'],
                 'Narxi':['9000','80000','65000']}) 
```

```python
B
```

```python
merged = pd.merge(A,B) # withoout parameter// by default inner join
```

```python
merged
```

```python
merger_outer = pd.merge(A,B, how = "outer") # outer will join everthing
```

```python
merger_outer
```

```python
merged_left = pd.merge(A,B,how ='left') # left join / takes all from left even not common
```

```python
merged_left
```

```python
merged_left
```

```python
merged_right=pd.merge(A,B, how='right')
```

```python
merged_right
```

```python
A = pd.DataFrame({'Transport_turi':['poeyezd','avtomobil','kema'],
                 'Tezligi':['70','80','90']})
```

```python
A
```

```python
B
```

```python
pd.merge(A,B) # NO COMMON COLUMNS
```

```python
merg_no_com = pd.merge(A,B,left_on='Transport_turi',right_on ='Transport')
```
