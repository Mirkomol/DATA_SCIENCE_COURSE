```python
import pandas as pd
```

```python
import numpy as np
import sqlite3 as sql
```

```python
df = pd.read_csv("praktikum_datasets/world_population.csv",index_col = "rank")
```

```python
df.head()
```

```python
db = sql.connect("newdb.db")
```

```python
cursor = db.cursor()
```

```python
query = """
CREATE TABLE IF NOT EXISTS worldpop(
rank INT,
country TEXT PRIMARY KEY,
density DOUBLE,
population DOUBLE NOT NULL,
area INT NOT NULL
);
"""
cursor.execute(query)
db.commit()
```

```python
uz = df[df.country=='Uzbekistan']
```

```python
uz
```

```python
rank  = uz.index.values[0]
country = uz.at[rank,'country']
density = uz.at[rank,'density']
population = uz.at[rank,'pop2021']
area = uz.at[rank,'area']

print(f"{rank}, {country},{density},{population},{area}")
```

```python
query = f"""
INSERT INTO worldpop VALUES ({rank},'{country}',{density},{population},{area})
"""

cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```

```python
db.close()
```

# between in and like


```python
query = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country = 'Uzbekistan' AND year BETWEEN 2005 AND 2010"
dfuz = pd.read_sql(query,db)
dfuz
```

```python
query = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country = 'Uzbekistan' AND year IN (1990,1995,2000,2005)"
dfuz = pd.read_sql(query,db)
dfuz
```

```python
# LIKE 

query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN(2019,2020,2021)"
pd.read_sql(query,db)
```
