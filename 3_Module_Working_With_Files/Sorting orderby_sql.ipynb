```python
import sqlite3 as sql
```

```python
db = sql.connect("chinook.db")
cursor = db.cursor()
```

```python
query ="SELECT CustomerId,Firstname,Lastname,Phone,Country  FROM CUSTOMERS" 
df = pd.read_sql(query,db,index_col = 'CustomerId')
```

```python
query ="SELECT CustomerId,Firstname,Lastname,Phone,Country  FROM CUSTOMERS ORDER BY Country" 
df = pd.read_sql(query,db,index_col = 'CustomerId')
```
