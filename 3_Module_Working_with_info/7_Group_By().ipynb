```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("praktikum_datasets/automobile_data.csv",index_col=0)
```

```python
df.head()
```

# Group by

```python
df.company.value_counts()
```

```python
df.groupby('company').mean()
```

```python
df.groupby('company').max()
```

```python
df.groupby('company').price.max() # only column
```

```python
df.groupby(['company','body-style']).price.mean()
```

```python
# agg multpil function
df.groupby('company').agg([min,max,np.mean])
```

```python
df.groupby('company').agg([min,max, lambda x:np.median(x)])
```
