```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(16,6))

plt.plot(x,y,linewidth = 3,linestyle=":",color = "b",marker='v')
plt.title("sin(x) function")
plt.xlabel("pi")
plt.ylabel("sin(x)")
plt.grid(color="k",linestyle="--") # setka
```

```python
x = np.arange(0,4 * np.pi,0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(12,6))

plt.plot(x,sin,
        linewidth=1.5,
        linestyle="-.",
        color="red",
        alpha=0.7,
        marker = "h")


plt.plot(x,cos,
        linewidth=1.5,
        linestyle=":",
        color="c",
        alpha=0.7,
        marker = "v")

plt.title("sin x ba cos x function")

plt.xlabel("po")

plt.legend(["sin,x","cos(x)"],loc=0, fontsize = "x-large")
plt.grid(color="tomato",alpha=0.5,linestyle="--")
plt.savefig("grafik1.png")
plt.show()
```
