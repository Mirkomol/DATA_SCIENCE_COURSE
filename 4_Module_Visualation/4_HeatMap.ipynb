```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("praktikum_datasets/museum_visitors.csv",index_col=0,parse_dates=[0],usecols=[0,1])
```

```python
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
del df["yead"]
```

```python
df2 = df.groupby(['month','year']).max()
```

```python
df2.unstack(level=1)
```

```python
plt.figure(figsize =(12,6))
sns.heatmap(data=df2.unstack(level=0),cmap="Greens",linewidths=1,square=True)
xticks_labels = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
plt.xticks(np.arange(12)+.5,labels=xticks_labels)
plt.show()
```
