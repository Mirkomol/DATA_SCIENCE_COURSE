```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("praktikum_datasets/merc.csv")
```

```python
df.head()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='mileage',y='price')
#sns.regplot(data=df,x='mileage',y='price')
plt.show()
```

# ScaterPlot_2

```python
df = df[df.year>2000]
plt.figure(figsize=(16,10))
sns.scatterplot(data=df,x='mileage',y='price',hue="fuelType", s=200)
#sns.regplot(data=df,x='mileage',y='price')
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```
