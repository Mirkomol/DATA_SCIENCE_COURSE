```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("praktikum_datasets/merc.csv")
df = df[df.year >=2010]
df = df[df.transmission!='other']
df = df[df.fuelType!='other']
df.head()
```

# Distribution


```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='price')
plt.grid()
plt.show()
```

```python
sns.displot(data=df,x='mileage', hue='transmission',multiple='stack', height = 5,aspect = 2)
plt.grid()
plt.show()
```

```python
sns.displot(data=df,x='price', col='transmission', height = 5,aspect = 2)
plt.show()
```

# KDE

```python
sns.displot(data=df,kind='kde',x='price', height=5,aspect=2)
plt.show()
```

```python
# kde true
sns.displot(data=df,kde=True,x='price', height=5,aspect=2)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df,x='price',shade=True)
```

# CDE

```python
# kde true
sns.displot(data=df, kind="ecdf", x='mileage', height=5,aspect=2)
plt.xticks(rotation = 90)
plt.grid(which='both')
plt.minorticks_on()
plt.show()
```
