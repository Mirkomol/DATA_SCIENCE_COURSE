```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("praktikum_datasets/merc.csv")
df = df[df.year >=2010]
df = df[df.transmission!='other']
df = df[df.fuelType!='other']
df.head()
```

```python
sns.histplot(data=df,x='price')
plt.show()
```

```python
sns.scatterplot(data=df, x='price',y='year',s=100)
```

```python
fig1 = plt.figure(figsize=(12,6))
```

```python
fig,ax = plt.subplots(2,1,figsize=(10,10))
sns.histplot(ax = ax[0], data=df,x='price')
sns.scatterplot(ax = ax[1],data=df, x='price',y='year',s=100)
plt.show()
```

```python
fig,ax = plt.subplots(2,2, figsize=(15,10))
fig.suptitle("grafik 2")
sns.histplot(ax = ax[0,0], data=df,x='price')
sns.scatterplot(ax = ax[0,1],data=df, x='price',y='year',s=100)
sns.scatterplot(ax=ax[1,0],data=df,x=df['year'],y=df['mileage'],hue='fuelType',s=50)
```
